```python
#Connect and stuff
sp = SPARQLWrapper('http://localhost:8080/sparql')
sp.setReturnFormat(JSON)
patientTerminationsQuery = """prefix dcterms: <http://purl.org/dc/terms/>
prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
prefix dataset: <http://purl.org/twc/semmdd/source/pican-wpic-pitt-edu/dataset/ppli-ssri/version/2012-09-08>
prefix e1: <http://purl.org/twc/semmdd/source/pican-wpic-pitt-edu/dataset/ppli-ssri/vocab/enhancement/1/>
prefix foaf: <http://xmlns.com/foaf/0.1/>
select ?patient
where 
{
?termentry dcterms:isReferencedBy dataset: .
?termentry e1:term \"0\" .
?termentry 
foaf:isPrimaryTopicOf ?patient
}"""

 
sp.setQuery(patientTerminationsQuery)

prop_term_raw = sp.query().convert()
```

```python
main_query = """
prefix completedpatientID: <http://purl.org/twc/semmdd/source/pican-wpic-pitt-edu/dataset/ppli-ssri/ppli-hams.xls/typed/patient/>
prefix datasetvocab: <http://purl.org/twc/semmdd/source/pican-wpic-pitt-edu/dataset/ppli-ssri/vocab/>
prefix datasetmeasurement: <http://purl.org/twc/semmdd/source/pican-wpic-pitt-edu/dataset/ppli-ssri/ppli-hams.xls/version/2012-09-08/>
prefix rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
prefix openvocab: <http://open.vocab.org/terms/>
prefix measurementanswer: <http://purl.org/twc/semmdd/source/pican-wpic-pitt-edu/dataset/ppli-ssri/vocab/enhancement/1/>
prefix foaf: <http://xmlns.com/foaf/0.1/>
prefix dcterms: <http://purl.org/dc/terms/>
prefix question: <http://purl.org/twc/semmdd/source/pican-wpic-pitt-edu/dataset/ppli-ssri/Q>

SELECT DISTINCT ?patient ?cdate ?column ?question ?answer 
    WHERE {
        GRAPH <http://pican-wpic-pitt-edu/ppli_hams> {
            ?Measurement rdf:type datasetvocab:Measurement .
            ?Measurement openvocab:csvCol ?column .
            ?Measurement measurementanswer:for_question ?question .
            FILTER ( ?question IN (""" + chosen_quests_string + """)) .
            ?Measurement measurementanswer:hasAnswer ?answer .
            ?Measurement foaf:isPrimaryTopicOf ?patient .
            FILTER ( ?patient IN (""" + prop_term_string + """)) .
            ?Measurement dcterms:created ?cdate
        } 
    }"""
sp.setQuery(main_query)

main_data = sp.query().convert()
```

```python
shaped_data = dict()

print len(main_data['results']['bindings'])

for i in main_data['results']['bindings']:
    patient_id = i['patient']['value'].split('/')[-1]
    response_date = datetime.date(*[int(j) for j in i['cdate']['value'].split('-')])
    question = i['question']['value'].split('/')[-1]
    question_slot = question_slots[question]
    answer = int(i['answer']['value'].split('/')[-1])
    normed_answer = answer / float(max_ham[question])
    
    # First setup a dictionary for the patient if one does not already exist
    if not patient_id in shaped_data:
        shaped_data[patient_id] = dict()
    
    # Then setup the list for the date in that patient's data, if it hasn't already been encountered
    if not response_date in shaped_data[patient_id]:
        shaped_data[patient_id][response_date] = [None for i in question_slots.keys()]
    
    # Then put the answer in the appropriate slot
    # But first, raise an error if it's already been filled... clearly something got fucked up somewhere
    if shaped_data[patient_id][response_date][question_slot] != None:
        print "Conflicting data. Abort."
        print "Patient ID %s's response to question %s on date %s has already been filled with value %d" (patient_id, question, response_date.ctime(), answer)
        break
    
    shaped_data[patient_id][response_date][question_slot] = normed_answer
```

```python
questions_missing = [[0,set([])] for i in chosen_quests]
counter = 0 # count the number of responses ever made
repeat_off = dict()

for i in shaped_data:
    repeat_off[i] = [0,set()]
    for date in shaped_data[i]:
        counter += 1
        for resp in range(len(shaped_data[i][date])):
            if resp == 24: continue # This is missing entirely we've concluded
            if shaped_data[i][date][resp] == None:
                repeat_off[i][0] += 1
                #print "Patient %s did not respond to question %d on %s" % (i, resp+1, date.ctime())
                questions_missing[resp][0] += 1
                if date not in repeat_off[i][1]:
                    repeat_off[i][1].add(date)
                if i not in questions_missing[resp][1]:
                    questions_missing[resp][1].add(i)

for ind, i in enumerate(questions_missing):
    if i[0] == 0: 
        continue
    print "Question %d:" % (ind+1)
    if i[0] == counter:
        print "This question is missing entirely"
        continue
    print "%d missing values from the following patients: %s" % (i[0], ','.join(i[1]))

for i in repeat_off:
    if repeat_off[i][0] == 0: 
        continue
    print "Patient %s is missing a response to %d question(s) on %d separate date(s)" % (i, repeat_off[i][0], len(repeat_off[i][1]))
```

```python
# For each question:
# For patients w/non-zero baseline:
# t-test between baseline and final response value


from scipy.stats.mstats import ttest_rel
# To start the t-test stuff, we need to, for each question, create two populations: one of patients' initial responses
# and one of patients' final responses. 

populations = {q:{'initial':[],'final':[],'useful':[]} for q in question_slots}

for patient in filtered_data:
    initial_session = filtered_data[patient][0]
    final_session = filtered_data[patient][-1]
    for question in range(len(initial_session)):
        quest_str = question_slots_r[question]
        if initial_session[question] == 0: # don't bother working with this
            continue
        populations[quest_str]['initial'].append(initial_session[question])
        populations[quest_str]['final'].append(final_session[question])
        populations[quest_str]['useful'].append(patient)
        
    
for q in populations:
    #populations[q]['initial'] = np.array(populations[q]['initial'])
    #populations[q]['final'] = np.array(populations[q]['final'])
    
    populations[q]['t'], populations[q]['p'] = ttest_rel(populations[q]['initial'],populations[q]['final'])
    
    #print "%s: t-statistic: %f p-value: %f: population size: %d" % (q, populations[q]['t'], populations[q]['p'], len(populations[q]['initial']))  
    
```

```python
from scipy.interpolate import UnivariateSpline

splined_data = dict()
spline_degree = 1
min_data = 3


for spline_degree in range(1,4):
    error = 0
    errcount = 0
    badcount = 0
    splined_data[spline_degree] = dict()
    for patient in filtered_by_time:
        insufficient_data = False
        x = np.array(filtered_by_time[patient]['days'])
        if len(x) <= min_data: # cubic spline
            #print "Not enough data for patient %s to use the cubic spline, only %d datapoints" % (patient, len(x))
            continue
        patient_data = []
        for question in range(len(chosen_quests)):
            y = np.array([i[question] for i in filtered_by_time[patient]['data']])
            good_indices = np.array([ind for ind, i in enumerate(y) if i != None])
            filtered_x = x[good_indices]
            filtered_y = y[good_indices]
            if len(filtered_y) <= spline_degree:
                insufficient_data = True
                break
            #TODO: handle len(filtered_x) <= spline_degree. Does not happen w/UPitt SSRI data, but could still happen.
            # Best idea: linear interpolate between adjacent points to replace None values.
            # Not worth the effort for UPitt PPLI SSRI, only 8 patients w/missing data.
            full_space = np.linspace(0,keep_months*31,num=keep_months*31)
            s = 1
            spl = UnivariateSpline(filtered_x,filtered_y,k=spline_degree,s=s)
            s_actual = spl.get_residual()
            knots = spl.get_knots()
            splined = spl(full_space)
            patient_data.append(splined)
            # now we're going to measure the error for this question and add it to the overall error
            ms = sqrt(sum((splined[filtered_x] - filtered_y)**2)/len(filtered_y))
            error += ms
            errcount += 1
        if insufficient_data:
            badcount += 1
            continue
        splined_data[spline_degree][patient] = np.array(patient_data).T
    error /= errcount
    print "Error for spline degree %d: %f" % (spline_degree, error)
    print "Patients we had to skip for this spline degree %d" % badcount

```

```python
# So, as a little benchmark, I'm going to take the average response for each question at each time and see how well that does
averages = dict()
for spline_degree in range(1,4):
    averages[spline_degree] = []
    average_patient = np.average(np.array(splined_data[spline_degree].values()),axis=0)
    averages[spline_degree].append(average_patient)
    # compute error by just measuring average abs difference for each point
    diff = np.zeros(average_patient.shape)
    for patient in splined_data[spline_degree]:
        diff += np.fabs(average_patient - splined_data[spline_degree][patient])
    diff /= len(splined_data)
    averages[spline_degree].append(diff)

```
